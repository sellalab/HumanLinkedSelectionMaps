```python
import seaborn
import numpy as np
import matplotlib as mpl

# set high DPI on figures globally
mpl.rcParams['figure.dpi'] = 200
plt = mpl.pyplot
```

```python
# set the population to load the data for
pop = 'MXL'

# open the zipped archive of data (sorted by distance to genes)
npz_archive = np.load('{}.cmtogene.binned.npz'.format(pop))

# keys for data archive 
# rhe/pon refer to rhesus and orangutan
# prefix x/a refer to x and autosome
archive_keys = 'xpi xrhe xpon xcnt api arhe apon acnt cm'.split()

# load each of the saved arrays
# pi, rhesus, orang, site count for x and aut, respectively and avg cM per bin
xpi, xrhe, xpon, xcnt, api, arhe, apon, acnt, cm = [npz_archive[k] for k in archive_keys]
```

```python
# plot pi values as function of distance from genes for X and Aut
plt.figure(figsize=(10,5))
plt.scatter(cm, api, color='mediumorchid', label='A')
plt.scatter(cm, xpi, color='darkorange', label='X')
plt.title('pi values')
plt.ylabel('pi')
plt.xlabel('cM to gene')

# sometimes scatter plot scaling is crap so i set them manually like this
plt.ylim(0.9 * xpi.min(), 1.1*api.max()
plt.legend()
plt.show()
```

```python
# plot pi values divided by divergence to rhesus macaque
plt.figure(figsize=(10,5))
plt.scatter(cm, api / arhe, color='mediumorchid', label='A')
plt.scatter(cm, xpi / xrhe, color='darkorange', label='X')

plt.title('pi corrected for rhesus divergence')
plt.ylabel('pi / D')
plt.xlabel('cM to gene')

plt.legend()
plt.show()
```

```python
# plot divergence to rhesus macaque
plt.figure(figsize=(10,5))
plt.scatter(cm, arhe, label='A', color='mediumorchid')
plt.scatter(cm, xrhe, label='X', color='darkorange')

plt.title('divergence to rhesus macaque')
plt.ylabel('D')
plt.xlabel('cM to gene')

plt.legend()
plt.show()
```
