```python
np.load('sortdists.npz')['sortdists']
np.load('sortpi.npz')['sortpi']
np.load('sortrhe.npz')['sortrhe']
np.load('sortpon.npz')['sortpon']
```

```python
xpi = np.load('pivals/chrX.pi.npy')
xrhe = np.load('dvals/rheMac3.chrX.d.npy')
xpon = np.load('dvals/ponAbe2.chrX.d.npy')
```

```python
cmbins = [1.40706680e-04,   7.13008262e-04,   2.48163423e-03, 7.73421029e-03,   2.22823615e-02,   5.90187610e-02, 1.29574320e-01,   2.61660955e-01,   5.30814331e-01, 4.33775062e+00]
```

```python
xdist = np.load('xmindists.npz')['mindists']
```

```python
xsidx = np.argsort(xdist)
```

```python
xdist, xpi, xrhe, xpon = [x[xsidx] for x in xdist, xpi, xrhe, xpon]
```

```python
xidx = np.searchsorted(xdist, cmbins)
```

```python
i = 0
avg_xpi = []
avg_xrhe = []
avg_xpon = []
for j in xidx:
    avg_xpi.append(xpi[i:j].mean())
    avg_xrhe.append(rhe[i:j].mean())
    avg_xpon.append(pon[i:j].mean())
    i = j
```

```python
avg_xpi, avg_xpon, avg_xrhe = [np.array(a) for a in avg_xpi, avg_xpon, avg_xrhe]
```

```python
step = len(pivals) / 10
avg_pi = []
avg_rhe = []
avg_pon = []
for i in xrange(0, len(pivals), step):
    avg_pi.append(pivals[i:i+step].mean())
    avg_rhe.append(rhe[i:i+step].mean())
    avg_pon.append(pon[i:i+step].mean())
```

```python
avg_pi, avg_pon, avg_rhe = [a[:10] for a in avg_pi, avg_pon, avg_rhe]
```

```python
avg_pi, avg_pon, avg_rhe = [np.array(a) for a in avg_pi, avg_pon, avg_rhe]
```

```python
# individual values
plt.figure(figsize=(10,5))
plt.scatter(np.log10(cmbins), avg_xpi / avg_pi)
plt.title('pi values')
plt.ylabel('X:A ratio')
plt.xlabel('log10 cM to gene')
plt.show()

plt.figure(figsize=(10,5))
plt.scatter(np.log10(cmbins), avg_xrhe / avg_rhe, label='macaque D values')
plt.title('D values')
plt.ylabel('X:A ratio')
plt.xlabel('log10 cM to gene')

plt.scatter(np.log10(cmbins), avg_xpon / avg_pon, label='orangutan D values')
plt.legend()
plt.show()
```

```python
# pi/d
plt.figure(figsize=(10,5))
plt.scatter(np.log10(cmbins), avg_xpi / avg_pi * avg_rhe / avg_xrhe, label='D macaque')
plt.title('pi/D values')
plt.ylabel('X:A pi / X:A D')
plt.xlabel('log10 cM to gene')
# plt.show()

# plt.figure(figsize=(10,5))
plt.scatter(np.log10(cmbins), avg_xpi / avg_pi * avg_pon / avg_xpon, label='D orangutan')
plt.legend()
plt.show()
```
